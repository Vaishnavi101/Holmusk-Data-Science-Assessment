```python
#Importing packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
from datetime import date
```

```python
#Reading the joined dataset
joineddata = pd.read_csv('data/joined_data.csv')
joineddata_original = joineddata
```

```python
joineddata
```

```python
#Find the percentage of missing values in each column
((joineddata.isnull().sum())/len(joineddata))*100
```

```python
#Check the data type of variables
joineddata.dtypes
```

```python
#Imputing missing values in 'medical_history_2' and 'medical_history_5' columns to 0 and changing to int type
joineddata['medical_history_2'] = joineddata['medical_history_2'].replace(np.nan, 0).astype(int)
joineddata['medical_history_5'] = joineddata['medical_history_5'].replace(np.nan, 0).astype(int)
```

```python
#Check the distribution of object variables
f, axes = plt.subplots(2, 2,figsize=(16,10))
sns.countplot(joineddata['medical_history_3'],ax=axes[0][0])
sns.countplot(joineddata['gender'],ax=axes[0][1])
sns.countplot(joineddata['race'],ax=axes[1][0])
sns.countplot(joineddata['resident_status'],ax=axes[1][1])
```

```python
#Replace 'yes:1' and 'no:0' in "medical_history_3" column.
joineddata = joineddata.replace({'medical_history_3' : { 'No' : 0, 'Yes' : 1, '0' : 0, '1' : 1 }})

#Convert gender to binary variable- "Female/F : 0", "Male/M : 1"
joineddata = joineddata.replace({'gender' : { 'Female' : 0, 'Male' : 1, 'f' : 0, 'm' : 1 }})

#Replace 'chinese : Chinese', 'India: Indian' in "race" column.
joineddata = joineddata.replace({'race' : { 'chinese' : 'Chinese', 'India' : 'Indian'}})

#Combine Singaporean and Singapore Citizen as 'Singapore Citizen' in "resident_status" column
joineddata = joineddata.replace({'resident_status' : { 'Singaporean' : 'Singapore citizen'}})
```

```python
#Check the distribution of continuous variables
f, axes = plt.subplots(1, 5,figsize=(19,5))
sns.distplot(joineddata['lab_result_1'],hist=False,ax=axes[0])
sns.distplot(joineddata['lab_result_2'],hist=False,ax=axes[1])
sns.distplot(joineddata['lab_result_3'],hist=False,ax=axes[2])
sns.distplot(joineddata['weight'],hist=False,ax=axes[3])
sns.distplot(joineddata['height'],hist=False,ax=axes[4])
```

```python
#Distribution of the target variable
plt.figure(figsize=(10,4))
plt.tight_layout()
sns.distplot(joineddata['amount'])
```

```python
#Statistical details of the targer variable
joineddata['amount'].describe()
```

```python
#Find the age of patient using the 'date_of_birth' column.
def calculate_age(born):
    born = datetime.strptime(born, "%Y-%m-%d").date()
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

cal_age = joineddata['date_of_birth'].apply(calculate_age)
joineddata.insert(loc=30, column='patient_age', value=cal_age)
```

```python
#Calculate 'no_days_admitted' from 'date_of_admission' and 'date_of_discharge'
joineddata['date_of_admission'] = pd.to_datetime(joineddata['date_of_admission'])
joineddata['date_of_discharge'] = pd.to_datetime(joineddata['date_of_discharge'])

cal_days = (joineddata['date_of_discharge'] - joineddata['date_of_admission']).dt.days
joineddata.insert(loc=3, column='no_days_admitted', value=cal_days)
```

```python
#Removing columns used in adding new features
joineddata = joineddata.drop(columns = ['date_of_admission', 'date_of_discharge', 'date_of_birth'])
```
